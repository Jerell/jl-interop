```julia
include("src/lib/fluids/Compositions.jl")
using .Compositions
using Plots; pythonplot()
using CoolProp, Unitful

p = range(1, 150, length=300)
t = range(-20, 120, length=300)

enthalpy(p, t) = PropsSI("H", "P", p * u"bar", "T", t * u"C", compositions.co2pure)

z = @. enthalpy(p', t)
contour(p, t, z)
```

```julia
using LaTeXStrings, Plots; pythonplot()

f(x, y) = (3x + y^2) * abs(sin(x) + cos(y))

x = range(0, 5, length=100)
y = range(0, 3, length=50)
z = @. f(x', y)
contour(x, y, z)
```

```julia
contourf(x, y, z, levels=20, color=:turbo)
title!(L"(3x + y^2)|\sin(x) + \cos(y)|")
xlabel!(L"x")
ylabel!(L"y")
```
